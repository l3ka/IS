```python
import pandas as pd
import numpy as np
```

```python
# import modules
import random
import datetime

# create function accepting a single parameter, the year as a four digit number
def get_random_date(year):

    # try to get a date
    try:
        return datetime.datetime.strptime('{} {} {} {} {} {}'.format(random.randint(1, 366),
                                                                  random.randint(1, 12),
                                                                  random.randint(0, 23),
                                                                  random.randint(0, 59),
                                                                  random.randint(0, 59),
                                                                  year),
                                                                  '%j %m %H %M %S %Y')

    # if the value happens to be in the leap year range, try again
    except ValueError:
        get_random_date(year)
```
